```python
# df_borough.loc[:, 'year'] = df_borough['date'].apply(lambda x: x.year)

# Get the average annual price over time by borough 
df_avg_price_by_year = df.groupby(['borough','geocode', 'year'], observed=True).agg({'avg_price': 'mean'}).reset_index()

print(df_avg_price_by_year.shape)
df_avg_price_by_year.head()
```

```python
print(df_avg_price_by_year.info())

df_avg_price_by_year['borough'] = df_avg_price_by_year['borough'].astype(str)
df_avg_price_by_year['geocode'] = df_avg_price_by_year['geocode'].astype(str)

print(df_avg_price_by_year.info())
```

```python
# # Check by plotting price over time for one borough

# selected_borough = 'Kensington & Chelsea'
# df_subset = df[df['borough'] == selected_borough].sort_values('date')

# df_subset.plot(kind='line', x='date', y='avg_price')
# plt.xlabel('Date')
# plt.ylabel('Average Price [M]')
# plt.title('{}: Average Price over Time'.format(selected_borough))
# plt.grid()
# plt.show()
```

```python
list_of_boroughs = list(df['borough'].unique())

dict_ratio = {}
for borough in list_of_boroughs:
    df_sub = df_avg_price_by_year[df_avg_price_by_year['borough']==borough]
    num = df_sub[df_sub['year']==2023]['avg_price'].values
    denom = df_sub[df_sub['year']==2003]['avg_price'].values
    dict_ratio[borough] = (num/denom)[0]
```

```python
max_val = max(dict_ratio.values())
max_key = max(dict_ratio, key=dict_ratio.get)
print(max_key, max_val)
```

```python
# Plot ratio by price
# del df_borough_2023
df_borough_2023 = df_avg_price_by_year[df_avg_price_by_year['year']==2023]
df_borough_2023.loc[:,'ratio'] = df_avg_price_by_year['borough'].apply(lambda x: dict_ratio[x])

df_borough_2023 = df_borough_2023.reset_index(drop=True)
print(df_borough_2023.head())
```
